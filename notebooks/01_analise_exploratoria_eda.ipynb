```python
# Célula 1: Importações e Carregamento
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os

# Adiciona a raiz do projeto ao path para podermos importar nossos próprios módulos
sys.path.append(os.path.abspath('..'))

from src.preprocessing import run_preprocessing_pipeline

# Carrega os dados usando a nossa própria função (reaproveitamento de código!)
df = run_preprocessing_pipeline('../data/BASE DE DADOS PEDE 2024 - DATATHON.xlsx')

print(f"Total de registros: {len(df)}")
df.head()
```

```python
# Célula 2: Distribuição do Target
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Risco_Defasagem', palette='Set2')
plt.title('Distribuição de Alunos: Risco de Defasagem (1) vs Sem Risco (0)')
plt.show()
```

```python
# Célula 3: Matriz de Correlação dos Indicadores Numéricos
plt.figure(figsize=(12, 8))
# Seleciona apenas colunas numéricas
cols_num = df.select_dtypes(include=['float64', 'int64']).columns
correlacao = df[cols_num].corr()

# Plota o mapa de calor
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Matriz de Correlação entre Indicadores e Risco de Defasagem')
plt.show()
```
